```python
import multiprocessing

from raims.data import load_msp_documents
from spec2vec.model_building import train_new_word2vec_model

defaults = {
    'iterations': list(range(0, 103, 2)),
    'learning_rate_initial': 0.01,
    'workers': multiprocessing.cpu_count() / 2,
    'progress_logger': True,
}
```

```python
mona_random_w2v_model = train_new_word2vec_model(
    documents=load_msp_documents('data/split/mona-random/train.msp'),
    filename= 'model/mona-random-w2v.model',
    **defaults
)
```
